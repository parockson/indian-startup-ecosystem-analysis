Problem Definition: The problem can be defined as analyzing the funding received by Indian start-ups from 2018 to 2021. This involves understanding the trends in funding, identifying the

## Hypotheses

Hypotheses:
1. The overall trend in funding received by Indian start-ups from 2018 to 2021 shows a steady increase, reflecting the growing confidence and interest of investors in the Indian start-up ecosystem.
2. The year 2021 saw the highest total funding amount, possibly due to increased digitization efforts, government support for entrepreneurship, and the emergence of new technologies and markets.
3. Noticeable fluctuations or anomalies in funding amounts across the four years may be attributed to economic factors, regulatory changes, global events, and shifts in investor sentiments towards certain industries or business models.
4. The top industries or sectors that attracted the most funding during this period include e-commerce, fintech, healthcare, and edtech, driven by consumer demand, technological advancements, and regulatory reforms.
5. Funding patterns differ among various stages of start-ups, with early-stage start-ups often focusing on product development and market validation, while growth-stage start-ups aim to scale operations and capture larger market shares.
6. Prominent investors participating in funding Indian start-ups during this period include venture capital firms, corporate investors, and angel investors, with preferences for innovative technologies, disruptive business models, and scalable solutions addressing critical market needs.
7. There is a positive correlation between the size of funding rounds and the subsequent success of start-ups, as larger funding rounds provide resources for expansion, product development, and market penetration, increasing the likelihood of achieving growth milestones and generating returns for investors.
8. Geographical trends indicate that funded start-ups are concentrated in major metropolitan areas such as Bangalore, Mumbai, and Delhi-NCR, where access to talent, infrastructure, and networking opportunities influences funding amounts and investor interest.
9. A significant percentage of funded start-ups from each year went on to achieve significant growth or exit events, such as acquisitions or IPOs, highlighting the maturity and potential of the Indian start-up ecosystem.
10. The COVID-19 pandemic impacted the funding landscape for Indian start-ups during these years by accelerating digital transformation, shifting consumer behaviors, and fostering innovation in sectors such as remote work solutions, healthtech, and online education. There may be notable shifts in investment strategies towards resilient business models, technology-enabled solutions, and sectors poised for long-term growth post-pandemic.
